```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
```

```python
def webtoon_order(info):
    url = f"https://comic.naver.com/webtoon/weekday?order={info}"
    return url
```

```python
print("인기순: User, 업데이트순 : Update, 조회순 : ViewCount, 별점순 : StarScore\n정렬방법을 입력해주세요:  ")
order = input()
```

```python
response = requests.get(webtoon_order(order), headers={"user-agent":"Mozilla/5.0"})
soup = bs(response.text,'html.parser')
webtoon_week = soup.select("div.col_inner") # 요일이 포함된 모든웹툰제목
week = ["월요일 웹툰", "화요일 웹툰", "수요일 웹툰", "목요일 웹툰", "금요일 웹툰", "토요일 웹툰", "일요일 웹툰"]
```

```python
webtoon_total_list = [] # beautifulSoup 상태를 리스트화 한전체웹툰제목
webtoon_day = [] # 웹뚠

# 웹툰제목을 리스트화
for y in webtoon_week:
    webtoon_total_list.append(y.get_text().split("\n"))

# 2차원을 1차원으로
null_remove = np.concatenate(webtoon_total_list).tolist()

# 빈값을 제거해서 새로운배열에 append
for x in null_remove: 
    if x != "" :
        webtoon_day.append(x)

# 딕셔너리에 요일넣어주기
webtoon_dict = {}
for x in range(len(webtoon_week)):
    webtoon_dict[week[x]] = {}

for x in range(len(webtoon_week)): # div.col_inner가 7개라서 7번 돌아갈거고
    if webtoon_week[x]: # week_web[0] > 월요일
        title = webtoon_week[x].select('a.title') # 월요일 웹툰제목 긁어왔어
        i = 0
        for y in title:
            i += 1
            webtoon_dict[week[x]][i] = y.get_text()
```

```python
df = pd.DataFrame(webtoon_dict)
# 요일별 연재 웹툰 개수가 달라서 각 요일의 최하단에 NaN값이 발생
df
```
